```python
def check_df(dataframe, head=5):
    print('\n\/ Форма ################################################################\n')
    print(dataframe.shape)
    print('\n\/ Типы данных ##########################################################\n')
    print(dataframe.dtypes)
    print('\n\/ Шапка датасета #######################################################\n')
    print(dataframe.head(head))
    print('\n\/ Пропущенные значения #################################################\n')
    print(dataframe.isnull().sum())
    print('\n\/ Распределение ########################################################\n')
    print(dataframe.describe([0, 0.05, 0.50, 0.95, 0.99, 1]).T)

check_df(df)
```

```python
data_pred = df
data_pred.head().T
```

### Закодируем фичи neighbourhood_group и neighbourhood labelencoder, а roomtype dummy

```python
le = preprocessing.LabelEncoder()

le.fit(data_pred['neighbourhood_group'])
data_pred['neighbourhood_group']=le.transform(data_pred['neighbourhood_group'])

le.fit(data_pred['neighbourhood'])
data_pred['neighbourhood']=le.transform(data_pred['neighbourhood'])
```

```python
data_pred = pd.get_dummies(data_pred)
```

```python
data_pred.head()
```

### Оценим качество простой линейной регрессии, RidgeCV, LassoCV и ElasticNetCV по метрикам R2, MAE, RMSE, MSE отталкиваясь от текущего состояния датасета

```python
X = data_pred.drop(['price'], inplace=False, axis=1)
y = data_pred['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
models = [('LR', LinearRegression()),
          ("RidgeCV", RidgeCV()),
          ("LassoCV", LassoCV()),
          ("ElasticNetCV", ElasticNetCV())]

rmse_scores = []
r2_scores = []
mae_scores = []
mse_scores = []
execution_times = []

for name, regressor in models:
    start_time = time.time()

    # Обучение модели
    regressor.fit(X_train, y_train)

    # Предсказание
    y_pred = regressor.predict(X_test)

    # RMSE
    rmse = np.mean(np.sqrt(-cross_val_score(regressor, X, y, cv=5, scoring="neg_mean_squared_error")))
    rmse_scores.append(rmse)
    
    # R^2 score
    r2 = metrics.r2_score(y_test, y_pred)
    r2_scores.append(r2)

    # MAE
    mae = metrics.mean_absolute_error(y_test, y_pred)
    mae_scores.append(mae)

    # MSE
    mse = metrics.mean_squared_error(y_test, y_pred)
    mse_scores.append(mse)

    # Время выполнения
    execution_time = time.time() - start_time
    execution_times.append(execution_time)

    # Оценка предсказания
    error = pd.DataFrame({'Actual Values': np.array(y_test).flatten(), 'Predicted Values': np.round(y_pred.flatten(), 2)})
   
    

    print(f"RMSE: {round(rmse, 4)} ({name})")
    print(f"R^2: {round(r2, 4)} ({name})")
    print(f"MAE: {round(mae, 4)} ({name})")
    print(f"MSE: {round(mse, 4)} ({name})")
    print(f"Время выполнения: {round(execution_time, 2)} seconds")
    print(f"Матрица предсказаний\n {error.head(10)}\n")

# Plot RMSE scores
plt.figure(figsize=(10, 10))
plt.bar([name for name, _ in models], rmse_scores)
plt.xlabel("Модель")
plt.ylabel("RMSE")
plt.title("Качество (RMSE)")
plt.show()

# Plot R^2 scores
plt.figure(figsize=(10, 10))
plt.bar([name for name, _ in models], r2_scores)
plt.xlabel("Модель")
plt.ylabel("R^2 Score")
plt.title("Качество (R^2 Score)")
plt.show()

# Plot MAE scores
plt.figure(figsize=(10, 10))
plt.bar([name for name, _ in models], mae_scores)
plt.xlabel("Модель")
plt.ylabel("MAE")
plt.title("Качество (MAE)")
plt.show()

# Plot MSE scores
plt.figure(figsize=(10, 10))
plt.bar([name for name, _ in models], mse_scores)
plt.xlabel("Модель")
plt.ylabel("MSE")
plt.title("Качество (MSE)")
plt.show()

# Plot execution times
plt.figure(figsize=(10, 10))
plt.bar([name for name, _ in models], execution_times)
plt.xlabel("Время выполнения (сек)")
plt.ylabel("Модель")
plt.title("Время выполнения для моделей")
plt.show()
```
